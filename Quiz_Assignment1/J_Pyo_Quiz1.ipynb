## Part2: Explorating Data Analysis (EDA)

### Data Set Information 

* carat: weight of the diamond(0.2 - 5.01).
one carat is equivalent to 0.2 grams.

* cut: quality of the cut(Fair, Good, Very Good, Premium, Ideal), so cut refers to how a rough diamond is shaped into a finshed diamond. 

* color: refers to the color of the diamond, Best color with clolorless. Worst color with a yellow tint. from J(worst) to D(best).  

* clarity: a measurement of how clear the diamond is (l1(worst),SI2, SI1, VS2, VS1, VVS2, VVS1, IF(best))

* depth: A diamond's depth is from top to bottom.. Total depth percentage : z/mean(x.y) = 2*z/(x+y)(43-79). 

* table:  A diamond's table is the flat, and need good balance. Wide of top of diamond relative to widest point(43-95)

* price: price in US dollars (\$326 - \$18,823)

* x: length in mm (0 - 10.74) of diamond

* y: width in mm (0 - 58.9) of diamond

* z: depth in mm (0 - 31.8) of diamond

## Part3 : Inferences

1. What proportion of diamonds are between 0.30 and 1.08 carat?
* Subset/Total = Answer

```python
#Calculate proprtion.
proportion = sum(df.carat.between(0.30,1.08) / df.carat.count())

#Print out with 2 decimal numbers
print(f"proportion of diamonds are between 0.30 and 1.08 is {proportion: .2f}")
```

2. How many of the diamonds have equal x and y dimensions?

```python
#loc get rows(and /or columns) with particular labels.
dimension = df.loc[(df.x!=0) & (df.y!=0), :]
dimension_xy = dimension[dimension.x == dimension.y].count()['x']

#print out
print(f"The number of the diamonds have equal x and y dimensions is {dimension_xy}")
```

3. How many of the diamonds have a carat value that is less than the mean carat value?

```python
less_carat = sum(df.carat < df.carat.mean())

#Print out the mean carat value
print(f"The mean carat value is {df.carat.mean() : .2f}")

#Print out the less than the mean carat value
print(f"The number of the diamonds have a carat value that is less than the mean carat value is {less_carat}")
```

4. How many diamonds have a Premium cut or better?
ascending order as follows: Fair/Good/Very Good/Premium/Ideal

```python
#Use the group by, and then make ascending order
group = df.groupby("cut").count()[["carat"]].rename({"carat":"count"},axis = "columns").sort_values(by="count")
display(group)

#Create barplot
new_barplot = sns.barplot(x ="cut", y= "count", data= group.reset_index())

#Print out barplot
print(new_barplot)
```

```python
PB_cut = sum(df.cut.isin(["Premium","Ideal"]))

#print out answer
print(f"There has {PB_cut:,} premium cut or better of diamonds")
```

5. Which diamond has the highest price per carat? What is its value?

```python
#Create a new field in df
new_field = df["perCarat"] = round(df.price / df.carat , 2)

#Use idmax() function to find the index of the maximum value along the index axis
max_index = df.perCarat.idxmax()

#Use max()
perCarat = df.perCarat.max()

#print out answer
print(f"Diamond index # {max_index} has the highest price per carat. The value of ${perCarat}")
```
